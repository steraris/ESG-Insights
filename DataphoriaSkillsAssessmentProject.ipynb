```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from DataImporter import DataImporter
from DataCleaner import DataCleaner
from DataTester import DataTester
```

## <font color=#009992>Greenhouse Gas Emissions (MtCO2e), Top 10 Countries by GPD, 2020</font>

```python
#Importing the csv file
data = pd.read_csv('historical_emissions.csv')
data = data.set_index('Country')
```

```python
data.head()
```

### <font color=#009992>Data Testing</font>

```python
tester = DataTester(data)
tester.test_null()
```

### <font color=#009992>Data Cleaning</font>

```python
cleaner = DataCleaner(data)
cleaner.replace_null('1990')
cleaner.drop_columns('Data source','Sector','Gas','Unit' )
```

### <font color=#009992>Slicing the required data for plotting</font>

```python
usa_data = data.loc['United States', '2019':'1990']
china_data = data.loc['China', '2019':'1990']
japan_data = data.loc['Japan', '2019':'1990']
germany_data = data.loc['Germany', '2019':'1990']
uk_data = data.loc['United Kingdom', '2019':'1990']
india_data = data.loc['India', '2019':'1990']
france_data = data.loc['France', '2019':'1990']
italy_data = data.loc['Italy', '2019':'1990']
canada_data = data.loc['Canada', '2019':'1990']
southkorea_data = data.loc['South Korea', '2019':'1990']
```

### <font color=#009992>Creating the Plot</font>

```python
plt.plot(usa_data.index[::-1], usa_data.values[::-1], color='#009992', label='USA')
plt.plot(china_data.index[::-1], china_data.values[::-1], color='#ff595e', label='China')
plt.plot(japan_data.index[::-1], japan_data.values[::-1], color='#ffca3a', label='Japan')
plt.plot(germany_data.index[::-1], germany_data.values[::-1], color='#8ac926', label='Germany')
plt.plot(uk_data.index[::-1], uk_data.values[::-1], color='#6a4c93', label='United Kingdom')
plt.plot(india_data.index[::-1], india_data.values[::-1], color='#1a281f', label='India')
plt.plot(france_data.index[::-1], france_data.values[::-1], color='#ce7b91', label='France')
plt.plot(italy_data.index[::-1], italy_data.values[::-1], color='#fffd82', label='Italy')
plt.plot(canada_data.index[::-1], canada_data.values[::-1], color='#7e6551', label='Canada')
plt.plot(southkorea_data.index[::-1], southkorea_data.values[::-1], color='#938581', label='South Korea')

plt.xlabel('Year')
plt.ylabel('Greenhouse Gas Emissions (MtCO2e)')
plt.legend()

# Set the xticks to show only every fifth year
plt.xticks(usa_data.index[::-5][::-1])

# Set the xlim to show only the range of years you want
plt.xlim(usa_data.index[-1], usa_data.index[0])

plt.show()
```

### <font color=#009992>Comments</font>

* The data visualization tracks CO2 emissions in countries with the largest GDP from 1990 to 2020

* China has quadrupled its emissions since 1990, becoming the largest contributor to global CO2 emissions

* The United States has remained a significant contributor to global emissions, with emissions remaining high and stable over time

* India's emissions have steadily increased, reflecting its growing economy and population

* The other seven countries included in the visualization have remained stable in their emissions with only minor fluctuations

* The visualization highlights the need for continued efforts to reduce emissions and address climate change, particularly in countries like China and India where emissions are growing rapidly.

## <font color=#009992>Organic Agriculture Land Use, Top 10 Countries by GPD, 2020</font>

### <font color=#009992>Importing the data</font>

```python
importer = DataImporter()
org = importer.read_csv('organic_data_2020.csv')
org.head()
```

### <font color=#009992>Data Testing/Cleaning</font>

```python
tester = DataTester(org)
tester.test_null()
```

```python
cleaner = DataCleaner(org)
cleaner.drop_columns('Domain Code','Domain','Area Code (M49)','Element Code','Element','Item Code','Item','Year Code','Year','Unit','Flag','Flag Description')
org['Area'] = org['Area'].replace('United Kingdom of Great Britain and Northern Ireland', 'UK')
org.head()
```

```python
# Filter the data to only include the countries and year we're interested in
countries = ["Canada", "China", "France", "Germany", "India", "Italy", "Japan", "Republic of Korea", "UK", "United States of America"]
mask = (org["Area"].isin(countries))
filtered_data = org.loc[mask]
```

### <font color=#009992>Creating the bar plot</font>

```python
plt.bar(filtered_data["Area"], filtered_data["Value"], color="#009992")
plt.xticks(rotation=90)
plt.title("Organic Agriculture Land Use in 2020")
plt.xlabel("Country")
plt.ylabel("Land Use (1000 ha)")
plt.show();
```

### <font color=#009992>Comments</font>

- The data visualization displays total hectares of land cultivated using organic methods in the ten countries with the biggest GPD, for 2020

- India, France, and China have relatively high numbers of organically cultivated land

- Japan, Korea, and the UK have almost no organically grown land

- The size of countries may impact the differences observed in organically cultivated land

- Factors such as cultural norms, land availability, and government policies could explain differences in organic farming practices

- The visualization highlights significant differences in organic farming practices among the ten countries. Further research could explore the reasons for these differences.

## <font color=#009992>Number of Democracies by Age, World, 2020</font>

### <font color=#009992>Importing the Data</font>

```python
importer = DataImporter()
dem = importer.read_csv('number-democracies-age-bmr.csv')
dem.head()
```

### <font color=#009992>Data Testing/Cleaning</font>

```python
tester = DataTester(dem)
tester.test_null()
```

```python
cleaner = DataCleaner(dem)
cleaner.drop_columns('Code')
```

```python
dem = dem[(dem['Year'] == 2020) & (dem['Entity'] == 'World')]

#extract the relevant columns
dem_age_groups = dem[['number_dem_18_bmr_owid', 'number_dem_30_bmr_owid', 'number_dem_60_bmr_owid',
'number_dem_90_bmr_owid', 'number_dem_91plus_bmr_owid']].sum()
non_dem = dem['number_nondem_bmr_owid'].sum()
```

```python
#define the labels and data for the plot
labels = ['Non-Democracies', '1-18', '19-30', '31-60', '61-90', '91+']
data = [non_dem] + list(dem_age_groups)
```

### <font color=#009992>Creating the Plot</font>

```python
#create the horizontal bar chart
fig, ax = plt.subplots()
ax.barh(labels, data, color="#009992")
#add a title and axis labels
ax.set_title('Number of Countries by Age Group and Regime Type (2020)')
ax.set_xlabel('Number of Countries')
ax.set_ylabel('Age Group')
#adjust the limits and ticks of the y-axis
ax.set_ylim(top=5.5, bottom=-0.5)
ax.set_yticks(range(6))
ax.invert_yaxis()
plt.show()
```

### <font color=#009992>Comments</font>

- The visualization displays the number of countries categorized by the age of their democracies for 2020

- More than 70 countries are non-democratic

- Most democracies are either 19-30 years old (around 35 countries) or 31-60 years old (around 40 countries)

- Less than 25 democracies are over 61 years old

- The visualization raises questions about the global state of democracy

- It emphasizes the need to promote democratic values and institutions worldwide

- The age of democracies varies widely across the world

- Advocating for democracy and promoting democratic values globally is crucial to ensure fair and representative political systems in every country
